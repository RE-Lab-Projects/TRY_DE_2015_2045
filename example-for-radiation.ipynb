## Plot of an example day to compare 1h, 15min and 1min temporal resolution

```python
import pandas as pd
```

```python
# Load example data: Test Reference Year of region 4, extreme summer, year 2045
data1 = pd.read_csv('3_processed-data/TRY_4_s_2045_1min.csv', index_col=0)
data15 = pd.read_csv('3_processed-data/TRY_4_s_2045_15min.csv', index_col=0)
```

```python
# Calculate global radiation
data1['global irradiation interpolated [W/m^2]'] = data1['direct irradiance [W/m^2]']+data1['diffuse irradiance [W/m^2]']
data1['global irradiation synthetic [W/m^2]'] = data1['synthetic global irradiance [W/m^2]']
data15['global irradiation interpolated [W/m^2]'] = data15['direct irradiance [W/m^2]']+data15['diffuse irradiance [W/m^2]']
data15['global irradiation synthetic [W/m^2]'] = data15['synthetic global irradiance [W/m^2]']
```

## Compare interpolated global radiation to synthetic radiation

### 1min temporal resolution

```python
# Overview over the whole year
data1[['global irradiation interpolated [W/m^2]','global irradiation synthetic [W/m^2]']].plot(alpha=0.5, xlabel='datetime', figsize=(20,7))
```

```python
# Example days in may
start=123*1440
end=127*1440
data1[['global irradiation interpolated [W/m^2]','global irradiation synthetic [W/m^2]','clear sky irradiance [W/m^2]']].plot(alpha=0.5, xlabel='datetime',xlim=(start,end), figsize=(20,7))
```

### 15min temporal resolution

```python
# Example days in may
start=123*96
end=127*96
data15[['global irradiation interpolated [W/m^2]','global irradiation synthetic [W/m^2]','clear sky irradiance [W/m^2]']].plot(alpha=0.5, xlabel='datetime',xlim=(start,end), figsize=(20,7))
```

### Histrogram

```python
# Histogram
data1[['global irradiation interpolated [W/m^2]','global irradiation synthetic [W/m^2]']][data1['global irradiation interpolated [W/m^2]'] > 5].plot(kind='hist', alpha=0.5, xlabel='global radiation [W/m^2]',figsize=(20,7))
```
